```python
def dropping_unwanted (data):
    black_list=['Airline', 'Source', 'Destination', 'Route', 'Dep_Time', 'Arrival_Time', 'Duration', 'Total_Stops', 'Additional_Info', 'Price', 'month', 'day', 'year']
    p = [feature for feature in data.columns if feature not in black_list]
    unwanted = data.drop(p, axis=1, inplace=True)
    return data
```

```python
duration = train_data['Duration']

for i in range (10683):
    if (len (duration[i].split()))!=2:
        if 'h' in duration[i]:
            duration[i] = duration[i].strip() + ' 0m'
        elif 'm' in duration[i]:
            duration[i] = '0h {}'.format(duration[i].strip())
            
dur_hour = []
dur_minutes = []

for i in range (10683):
    dur_hour.append(int(duration[i].split()[0][:-1]))
    dur_minutes.append(int(duration[i].split()[1][:-1]))
    
train_data['hour'] = dur_hour
train_data['minutes'] = dur_minutes

train_data.drop('Duration', axis = 1, inplace= True)  
```
